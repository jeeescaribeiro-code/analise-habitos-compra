> Os modelos de Machine Learning utilizados têm como objetivo identificar padrões nos dados e gerar previsões confiáveis. O principal algoritmo aplicado foi o Random Forest, que combina múltiplas árvores de decisão para aumentar a precisão e reduzir erros. Para lidar com o problema de dados desbalanceados, foi utilizado o SMOTE, uma técnica que gera exemplos sintéticos da classe minoritária, permitindo que o modelo aprenda melhor padrões menos frequentes. Além disso, foi utilizado o Random Forest Regressor para prever valores numéricos contínuos, como renda ou faturamento. A validação do modelo foi feita por meio de cross-validation 5-fold, que divide os dados em partes e realiza múltiplos testes para garantir uma avaliação mais estável e confiável. Por fim, foi desenvolvido um simulador de predição, que permite inserir novos dados e obter previsões do modelo na prática, tornando a solução aplicável em contextos reais.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

sns.set_theme(style='whitegrid', palette='pastel')
plt.rcParams['figure.figsize'] = (10, 5)
plt.rcParams['axes.titlesize'] = 13

print('Bibliotecas carregadas com sucesso!')
```

```python
dados = pd.read_excel('/content/drive/MyDrive/Colab-Looker/habitos de compra.xlsx')
dados = dados.dropna(subset=[dados.columns[0]])  # remove as 3 linhas completamente vazias
print(f'Dataset carregado: {dados.shape[0]} linhas, {dados.shape[1]} colunas')
```

```python
# Renomear colunas para facilitar o uso ao longo do notebook
dados = dados.rename(columns={
    'Qual sua faixa etária?': 'FAIXA_ETARIA',
    'Qual é o seu gênero?': 'GENERO',
    'Qual sua principal ocupação?': 'OCUPACAO',
    'Qual sua faixa salarial (renda mensal)?': 'FAIXA_SALARIAL',
    'Com que frequência você realiza compras em supermercados?': 'FREQUENCIA_COMPRAS',
    'Normalmente, você faz compras:': 'TIPO_COMPRA',
    'Você costuma comparar preços entre supermercados antes de comprar?': 'COMPARA_PRECOS',
    '[Preço dos produtos]': 'FATOR_PRECO',
    '[Distância do supermercado]': 'FATOR_DISTANCIA',
    '[Promoções e ofertas]': 'FATOR_PROMOCOES',
    '[Variedade de produtos]': 'FATOR_VARIEDADE',
    '[Atendimento]': 'FATOR_ATENDIMENTO',
    '[Facilidade de estacionamento]': 'FATOR_ESTACIONAMENTO',
    'Você sente dificuldade em comparar preços entre diferentes supermercados?': 'DIFICULDADE_COMPARAR',
    'Você já utilizou ou utiliza algum aplicativo para comparar preços de supermercado?': 'USA_APP',
    'Se sim, qual(is)?': 'APP_QUAL',
    'Você consideraria útil um aplicativo que compara preços em supermercados e mostra qual é a opção mais vantajosa com base no valor total do carrinho e na distância da sua casa?': 'CONSIDERA_UTIL',
    '[Comparação de preços em tempo real]': 'FUNC_COMPARACAO',
    '[Cálculo da distância até o supermercado]': 'FUNC_DISTANCIA',
    '[Alerta de promoções e ofertas]': 'FUNC_ALERTAS',
    '[Planejamento da lista de compras]': 'FUNC_LISTA',
    '[Sugestão de mercado com melhor custo-benefício]': 'FUNC_SUGESTAO',
    '[Histórico de compras e controle de gastos]': 'FUNC_HISTORICO',
    'Você estaria disposto(a) a pagar por funcionalidades premium do aplicativo, como por exemplo: acesso antecipado a promoções, cupons de desconto, entre outros?': 'PAGARIA_PREMIUM',
    'Qual preço estaria disposto em pagar pelas funcionalidades premium?': 'PRECO_PREMIUM',
    'Você aceitaria ver anúncios no aplicativo em troca de acesso gratuito?': 'ACEITA_ANUNCIOS',
    '[Instagram]': 'CANAL_INSTAGRAM',
    '[WhatsApp]': 'CANAL_WHATSAPP',
    '[Facebook]': 'CANAL_FACEBOOK',
    '[TikTok]': 'CANAL_TIKTOK',
    '[E-mail]': 'CANAL_EMAIL',
    '[Indicação de amigos]': 'CANAL_INDICACAO',
    'Que tipo de atendimento você esperaria encontrar no aplicativo?': 'TIPO_ATENDIMENTO',
    'Como você prefere navegar em aplicativos de compras ou finanças?': 'PREFERENCIA_NAVEGACAO',
    'Quais benefícios você mais espera de um aplicativo?': 'BENEFICIOS_ESPERADOS',
    '[Dificuldade de uso]': 'BARREIRA_USO',
    '[Falta de confiança nas informações]': 'BARREIRA_CONFIANCA',
    '[Não achar útil para minha rotina]': 'BARREIRA_UTILIDADE',
    ' [Preferência por compras presenciais]': 'BARREIRA_PRESENCIAL'
})

# Converter fatores textuais para escala numerica (1-3-5)
relevancia_map = {'Pouco relevante': 1, 'Relevante': 3, 'Muito relevante': 5}
colunas_fatores = ['FATOR_PRECO','FATOR_DISTANCIA','FATOR_PROMOCOES',
                   'FATOR_VARIEDADE','FATOR_ATENDIMENTO','FATOR_ESTACIONAMENTO']
for col in colunas_fatores:
    dados[col] = dados[col].map(relevancia_map)

colunas_funcoes   = ['FUNC_COMPARACAO','FUNC_DISTANCIA','FUNC_ALERTAS',
                     'FUNC_LISTA','FUNC_SUGESTAO','FUNC_HISTORICO']
colunas_canais    = ['CANAL_INSTAGRAM','CANAL_WHATSAPP','CANAL_FACEBOOK',
                     'CANAL_TIKTOK','CANAL_EMAIL','CANAL_INDICACAO']
colunas_barreiras = ['BARREIRA_USO','BARREIRA_CONFIANCA','BARREIRA_UTILIDADE','BARREIRA_PRESENCIAL']
funcionalidades   = colunas_funcoes  # alias para compatibilidade

for col in colunas_funcoes + colunas_canais:
    dados[col] = pd.to_numeric(dados[col], errors='coerce')

print('Colunas renomeadas e tipos convertidos!')
dados.columns.tolist()
```

### Tratamento de Valores Faltantes

```python
print('Valores nulos por coluna:')
print(dados.isnull().sum()[dados.isnull().sum() > 0])
```

```python
# Preencher genero nulo com 'Prefiro nao informar'
print('Antes:', dados['GENERO'].value_counts(dropna=False).to_dict())
dados['GENERO'] = dados['GENERO'].fillna('Prefiro não informar')
print('Depois:', dados['GENERO'].value_counts(dropna=False).to_dict())
```

```python
# Preencher demais categoricas com 'Nao informado'
dados['FAIXA_ETARIA']= dados['FAIXA_ETARIA'].fillna('Não informado')
dados['FAIXA_SALARIAL']= dados['FAIXA_SALARIAL'].fillna('Não informado')

# Preencher notas numericas com a mediana de cada coluna
for col in colunas_funcoes + colunas_fatores + colunas_canais:
    dados[col] = dados[col].fillna(dados[col].median())

print('Valores nulos restantes:', dados.isnull().sum().sum())
```

De forma geral, os resultados mostram quais funcionalidades são mais bem aceitas e quais demandam melhorias para aumentar a satisfação dos usuários.



```python
col_analise = 'FUNC_COMPARACAO'
q1 = dados[col_analise].quantile(0.25)
q3 = dados[col_analise].quantile(0.75)
iqr = q3 - q1
lim_superior = q3 + (1.5 * iqr)
lim_inferior = q1 - (1.5 * iqr)

print(f'Q1: {q1}  |  Q3: {q3}  |  IQR: {iqr}')
print(f'Limite superior: {lim_superior}  |  Limite inferior: {lim_inferior}')
outliers = dados[(dados[col_analise] > lim_superior) | (dados[col_analise] < lim_inferior)]
print(f'Outliers em {col_analise}: {len(outliers)}')
```

# PARTE 3: Insights
## Rankings, Correlações, Cruzamentos e Testes

### 🔍 Ranking de Prioridades do Consumidor

```python
dados['score_exigencia_mercado']= dados[colunas_fatores].mean(axis=1)
dados['score_interesse_app']= dados[colunas_funcoes].mean(axis=1)
dados['score_geral']= (dados['score_exigencia_mercado'] + dados['score_interesse_app']) / 2
```

```python
ranking_fatores = dados[colunas_fatores].mean().sort_values(ascending=False)
ranking_funcoes = dados[colunas_funcoes].mean().sort_values(ascending=False)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].barh(ranking_fatores.index[::-1], ranking_fatores.values[::-1],
             color=sns.color_palette('Blues_r', len(ranking_fatores)))
axes[0].set_title('O que mais importa na escolha do supermercado')
axes[0].set_xlabel('Importância Média (1–5)'); axes[0].set_xlim(0, 5.5)
for i, v in enumerate(ranking_fatores.values[::-1]):
    axes[0].text(v + 0.1, i, f'{v:.2f}', va='center', fontsize=9)

axes[1].barh(ranking_funcoes.index[::-1], ranking_funcoes.values[::-1],
             color=sns.color_palette('Greens_r', len(ranking_funcoes)))
axes[1].set_title('Funcionalidades mais desejadas no app')
axes[1].set_xlabel('Importância Média (1–5)'); axes[1].set_xlim(0, 5.5)
for i, v in enumerate(ranking_funcoes.values[::-1]):
    axes[1].text(v + 0.1, i, f'{v:.2f}', va='center', fontsize=9)

plt.tight_layout(); plt.show()
```

> Em relação aos fatores de escolha, o preço se destaca como o mais importante, seguido por distância e promoções, indicando que os consumidores priorizam economia e conveniência. Já aspectos como variedade, estacionamento e principalmente atendimento aparecem com menor relevância, sugerindo que têm menos peso na decisão final.

>No que diz respeito às funcionalidades do aplicativo, a comparação de preços é a mais valorizada, seguida por sugestões e alertas, o que reforça o foco do usuário em economizar e tomar decisões mais informadas. Funcionalidades como histórico, distância e lista de compras também são consideradas importantes, mas com menor prioridade relativa. De forma geral, os resultados indicam que tanto na escolha do supermercado quanto no uso do aplicativo, os usuários estão fortemente orientados por preço e praticidade, destacando a importância de ferramentas que facilitem a comparação e a economia.

### Score de Exigência por Pessoa

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
configs = [
    ('score_exigencia_mercado', 'Exigência no Mercado', 'steelblue'),
    ('score_interesse_app',     'Interesse no App',     'seagreen'),
    ('score_geral',             'Score Geral',          'coral'),
]
for ax, (col, titulo, cor) in zip(axes, configs):
    ax.hist(dados[col].dropna(), bins=14, color=cor, edgecolor='white', alpha=0.85)
    media = dados[col].mean()
    ax.axvline(media, color='black', linestyle='--', linewidth=1.5, label=f'Média: {media:.2f}')
    ax.set_title(titulo); ax.set_xlabel('Score (1–5)'); ax.legend(fontsize=9)

plt.suptitle('Distribuição dos Scores por Respondente', fontsize=13)
plt.tight_layout(); plt.show()
```

A exigência no mercado apresenta uma distribuição relativamente equilibrada, com concentração em valores médios-altos, sugerindo que os usuários têm um padrão razoavelmente elevado ao escolher serviços. Já o interesse no app mostra maior dispersão, com presença de notas mais baixas, mas também uma concentração significativa nas faixas mais altas, indicando que, embora nem todos estejam engajados, há um grupo relevante com alto interesse. ### Correlação

```python
colunas_num = colunas_fatores + colunas_funcoes
corr = dados[colunas_num].corr()
mask = np.triu(np.ones_like(corr, dtype=bool))

plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm',
            mask=mask, linewidths=0.5, vmin=-1, vmax=1, annot_kws={'size': 9})
plt.title('Mapa de Correlação: Fatores × Funcionalidades', fontsize=14)
plt.xticks(rotation=40, ha='right')
plt.tight_layout(); plt.show()
```

> De forma geral, observa-se que as correlações entre os fatores de decisão (preço, distância, promoções, etc.) e as funcionalidades do app são fracas ou próximas de zero, indicando que não há uma relação direta forte entre o que o usuário valoriza no supermercado e o interesse por funcionalidades específicas do aplicativo.

> Por outro lado, entre as funcionalidades, há correlações moderadas a fortes, destacando-se principalmente a relação entre comparação de preços e sugestão (0,92), além de correlações relevantes entre comparação e alertas, histórico e lista. Isso sugere que usuários que valorizam uma funcionalidade tendem a valorizar outras relacionadas, indicando um comportamento consistente de uso dentro do app.

```python
corr_pairs = (
    corr.where(~mask).stack().reset_index()
    .rename(columns={'level_0':'var1','level_1':'var2',0:'correlacao'})
    .assign(abs_corr=lambda x: x['correlacao'].abs())
    .sort_values('abs_corr', ascending=False).head(10)
)
print('Top 10 Correlações Mais Fortes:')
print(corr_pairs[['var1','var2','correlacao']].to_string(index=False))
```

```python
fig, axes = plt.subplots(1, 2, figsize=(13, 5))
r1 = dados['FATOR_PRECO'].corr(dados['FATOR_PROMOCOES'])
axes[0].scatter(dados['FATOR_PRECO'], dados['FATOR_PROMOCOES'], alpha=0.5, color='steelblue', edgecolors='white')
axes[0].set_xlabel('Importância: Preço'); axes[0].set_ylabel('Importância: Promoções')
axes[0].set_title(f'Preço vs. Promoções  (r = {r1:.2f})')
axes[0].annotate(f'r = {r1:.2f}', xy=(0.05, 0.92), xycoords='axes fraction', fontsize=11, color='red')

r2 = dados['score_interesse_app'].corr(dados['FUNC_COMPARACAO'])
axes[1].scatter(dados['score_interesse_app'], dados['FUNC_COMPARACAO'], alpha=0.5, color='coral', edgecolors='white')
axes[1].set_xlabel('Score Interesse no App'); axes[1].set_ylabel('Interesse: Comparação de Preços')
axes[1].set_title(f'Score App vs. Comparar Preços  (r = {r2:.2f})')
axes[1].annotate(f'r = {r2:.2f}', xy=(0.05, 0.92), xycoords='axes fraction', fontsize=11, color='red')

plt.tight_layout(); plt.show()
```

> “Score App vs. Comparar Preços”, há uma correlação positiva muito forte (r = 0,91), mostrando que quanto maior o interesse no uso do aplicativo, maior também é o interesse em comparar preços. Isso sugere um comportamento consistente: usuários mais engajados com o app tendem fortemente a utilizá-lo como ferramenta de comparação, evidenciando um padrão claro de uso orientado à busca por melhores ofertas.

```python
from scipy.stats import chi2_contingency

def cramer_coeficiente(coluna1, coluna2):
    tabela_cruzada = np.array(pd.crosstab(coluna1, coluna2))
    chi2 = chi2_contingency(tabela_cruzada)[0]
    soma = np.sum(tabela_cruzada)
    mini = min(tabela_cruzada.shape) - 1
    return np.sqrt(chi2 / (soma * mini))

print('V de Cramer: Gênero × Comparar Preços:')
print(f'  {cramer_coeficiente(dados["GENERO"], dados["COMPARA_PRECOS"]):.3f}')
print('  (0 = sem relação | 1 = relação perfeita)')
```

> O resultado obtido foi 0,162, indicando uma associação fraca entre gênero e o hábito de comparar preços. Isso sugere que o gênero não é um fator determinante relevante para explicar o comportamento de comparação de preços entre os indivíduos analisados.

### Testes Estatísticos

```python
from scipy.stats import ttest_ind, f_oneway

grupo_sim = dados[dados['DIFICULDADE_COMPARAR'] == 'Sim']['score_interesse_app'].dropna()
grupo_nao = dados[dados['DIFICULDADE_COMPARAR'] == 'Não']['score_interesse_app'].dropna()
stat, pvalue = ttest_ind(grupo_sim, grupo_nao)

print('T-Test: Score de Interesse no App: Dificuldade Sim vs. Não')
print(f'  Média (Sim): {grupo_sim.mean():.2f}  |  Média (Não): {grupo_nao.mean():.2f}')
print(f'  Estatística t: {stat:.3f}  |  p-valor: {pvalue:.4f}')
print('  ✅ Diferença significativa' if pvalue < 0.05 else '  ⚠️ Diferença não significativa')
```

> Isso significa que não há evidência suficiente para afirmar que a dificuldade impacta de fato o interesse no app. Em termos práticos, a variação observada pode ser apenas fruto de aleatoriedade da amostra, e não de uma relação consistente entre as variáveis.

```python
grupos = [dados[dados['FAIXA_SALARIAL'] == s]['score_interesse_app'].dropna()
          for s in dados['FAIXA_SALARIAL'].dropna().unique()]
grupos = [g for g in grupos if len(g) > 1]
f_stat, p_anova = f_oneway(*grupos)

print('ANOVA: Score de Interesse no App × Faixa Salarial')
print(f'  F: {f_stat:.3f}  |  p-valor: {p_anova:.4f}')
print('  ✅ Diferença significativa entre grupos de renda' if p_anova < 0.05 else '  ⚠️ Sem diferença significativa')
```

> A faixa salarial não parece influenciar o interesse no app, sugerindo que esse interesse é relativamente homogêneo entre pessoas com diferentes níveis de renda.

# PARTE 5: Machine Learning
## Predição de Comportamento do Usuário

<div style="background: #fff3e0; border-left: 5px solid #ff9800; padding: 18px 22px; border-radius: 8px; margin: 5px 0;">

**Três modelos preditivos:**

| Modelo | Pergunta | Tipo | Algoritmo |
|--------|----------|------|-----------|
| 1 | Vai **usar o app**? | Classificação | Random Forest + SMOTE |
| 2 | Vai **pagar o premium**? | Classificação | Random Forest |
| 3 | **Quanto** pagaria? | Regressão | Random Forest Regressor |

> ⚠️ **Nota sobre o dataset:** 98 amostras é um número pequeno para ML. Os resultados são bons para uma prova de conceito e análise exploratória. Com mais dados, os modelos ficam mais robustos.

```python
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import (classification_report, confusion_matrix,
                              ConfusionMatrixDisplay, mean_absolute_error, r2_score)
from imblearn.over_sampling import SMOTE

# Variáveis alvo
dados['TARGET_USA_APP'] = (dados['USA_APP'] == 'Sim').astype(int)
dados['TARGET_PAGARIA'] = (dados['PAGARIA_PREMIUM'] == 'Sim').astype(int)

preco_map = {
    'Não pagaria': 0,
    'Entre R$ 5,00 e R$ 10,00': 7.5,
    'Entre R$10,00 e R$20,00': 15,
    'Entre R$20,00 e R$30,00': 25
}
dados['TARGET_VALOR'] = dados['PRECO_PREMIUM'].map(preco_map)

# Features
features_cat = ['FAIXA_ETARIA','GENERO','FAIXA_SALARIAL','FREQUENCIA_COMPRAS',
                'TIPO_COMPRA','COMPARA_PRECOS','DIFICULDADE_COMPARAR','ACEITA_ANUNCIOS']
features_num = colunas_fatores + colunas_funcoes + ['score_geral','score_interesse_app']

df_ml = dados[features_cat + features_num +
              ['TARGET_USA_APP','TARGET_PAGARIA','TARGET_VALOR']].copy()
df_ml[features_cat] = df_ml[features_cat].fillna('Desconhecido')
df_ml[features_num] = df_ml[features_num].fillna(df_ml[features_num].median())

le_dict = {}
for col in features_cat:
    le = LabelEncoder()
    df_ml[col + '_enc'] = le.fit_transform(df_ml[col])
    le_dict[col] = le

features_enc = [c + '_enc' for c in features_cat] + features_num
print(f'Features: {len(features_enc)} | Amostras: {len(df_ml)}')
print(f'Target USA_APP:  {df_ml["TARGET_USA_APP"].value_counts().to_dict()}')
print(f'Target PAGARIA:  {df_ml["TARGET_PAGARIA"].value_counts().to_dict()}')
print(f'Target VALOR:    {df_ml["TARGET_VALOR"].value_counts().to_dict()}')
```

### 🤖 Modelo 1 — Vai usar o app? (Classificação + SMOTE)

```python
X = df_ml[features_enc]
y = df_ml['TARGET_USA_APP']

# SMOTE para balancear classes (86 nao usam x 12 usam)
sm = SMOTE(random_state=42, k_neighbors=5)
X_res, y_res = sm.fit_resample(X, y)
print(f'Amostras após SMOTE: {y_res.value_counts().to_dict()}')

X_train, X_test, y_train, y_test = train_test_split(
    X_res, y_res, test_size=0.25, random_state=42, stratify=y_res)

clf_app = RandomForestClassifier(n_estimators=200, max_depth=6, random_state=42)
clf_app.fit(X_train, y_train)
y_pred = clf_app.predict(X_test)

print('\nMODELO 1: Vai usar o app?')
print(classification_report(y_test, y_pred, target_names=['Não usa','Usa']))
cv = cross_val_score(clf_app, X_res, y_res, cv=5, scoring='f1')
print(f'F1 médio cross-val: {cv.mean():.2f} ± {cv.std():.2f}')
```

### O uso da técnica SMOTE eliminou o desequilíbrio dos dados originais, permitindo que o algoritmo aprendesse com precisão as características do grupo minoritário (os usuários reais).

### **Há uma Precisão de 94% para quem usa o app. Isso significa que o modelo é  assertivo e tende a não comete "falsos positivos", tornando-o ideal para campanhas de marketing, onde o objetivo é não desperdiçar recursos com o público errado.**

### Embora o modelo seja um pouco mais conservador na captura do total de usuários (Recall de 76%), a estabilidade demonstrada na validação cruzada (93%) confirma que ele é consistente e está pronto para uso prático.

```python
fig, axes = plt.subplots(1, 2, figsize=(13, 5))
ConfusionMatrixDisplay(confusion_matrix(y_test, y_pred),
                       display_labels=['Não usa App','Usa App']
                      ).plot(ax=axes[0], colorbar=False, cmap='Blues')
axes[0].set_title('Matriz de Confusão: Usa App?')

pd.Series(clf_app.feature_importances_, index=features_enc
         ).sort_values(ascending=False).head(10)[::-1
         ].plot(kind='barh', ax=axes[1], color='steelblue')
axes[1].set_title('Top 10 Features — Usar App'); axes[1].set_xlabel('Importância')
plt.tight_layout(); plt.show()
```

> O modelo nos diz que o seu produto não é apenas sobre "preço baixo", mas sobre facilitar um processo difícil (a comparação).

> Para converter novos usuários, sua comunicação não deve focar apenas em "economize dinheiro", mas sim em "pare de ter trabalho comparando preços sozinho". O modelo provou que a "Dificuldade de Comparar" é o que mais empurra as pessoas para dentro do seu app.

### Modelo 2: Vai pagar pelo premium? (Classificação)

```python
y2 = df_ml['TARGET_PAGARIA']
X2_train, X2_test, y2_train, y2_test = train_test_split(
    X, y2, test_size=0.25, stratify=y2, random_state=42)

clf_premium = RandomForestClassifier(n_estimators=200, max_depth=6, random_state=42)
clf_premium.fit(X2_train, y2_train)
y2_pred = clf_premium.predict(X2_test)

print('MODELO 2 — Vai pagar pelo premium?')
print(classification_report(y2_test, y2_pred, target_names=['Não paga','Paga']))
cv2 = cross_val_score(clf_premium, X, y2, cv=5, scoring='f1')
print(f'F1 médio cross-val: {cv2.mean():.2f} ± {cv2.std():.2f}')
```

###O modelo de previsão de pagamento não é confiável. Enquanto o uso do aplicativo é guiado por comportamentos lógicos (como a dificuldade de comparar preços), a intenção de pagar pelo Premium parece ser subjetiva ou influenciada por fatores que os dados atuais não capturam totalmente.

### Modelo 3: Quanto pagaria? (Regressão)

```python
df_reg = df_ml.dropna(subset=['TARGET_VALOR'])
X3, y3 = df_reg[features_enc], df_reg['TARGET_VALOR']

X3_train, X3_test, y3_train, y3_test = train_test_split(X3, y3, test_size=0.25, random_state=42)

reg = RandomForestRegressor(n_estimators=200, max_depth=5, random_state=42)
reg.fit(X3_train, y3_train)
y3_pred = reg.predict(X3_test)

mae = mean_absolute_error(y3_test, y3_pred)
cv3 = -cross_val_score(reg, X3, y3, cv=5, scoring='neg_mean_absolute_error')

print('MODELO 3 — Quanto pagaria?')
print(f'MAE no teste:            R$ {mae:.2f}')
print(f'MAE cross-val (5-fold):  R$ {cv3.mean():.2f} ± R$ {cv3.std():.2f}')
print()
print(pd.DataFrame({'Real (R$)': y3_test.values[:10],
                    'Previsto (R$)': y3_pred[:10].round(2)}).to_string(index=False))
```

```python
def prever_novo_usuario(
    faixa_etaria='25 a 34 anos', genero='Feminino',
    faixa_salarial='Entre 1 a 2 salários', frequencia_compras='1 vez por semana',
    tipo_compra='Presencialmente', compara_precos='Às vezes',
    dificuldade_comparar='Sim', aceita_anuncios='Sim',
    fator_preco=5, fator_distancia=3, fator_promocoes=4,
    fator_variedade=3, fator_atendimento=2, fator_estacionamento=2,
    func_comparacao=5, func_distancia=4, func_alertas=4,
    func_lista=3, func_sugestao=4, func_historico=3
):
    row = {
        'FAIXA_ETARIA': faixa_etaria, 'GENERO': genero,
        'FAIXA_SALARIAL': faixa_salarial, 'FREQUENCIA_COMPRAS': frequencia_compras,
        'TIPO_COMPRA': tipo_compra, 'COMPARA_PRECOS': compara_precos,
        'DIFICULDADE_COMPARAR': dificuldade_comparar, 'ACEITA_ANUNCIOS': aceita_anuncios,
        'FATOR_PRECO': fator_preco, 'FATOR_DISTANCIA': fator_distancia,
        'FATOR_PROMOCOES': fator_promocoes, 'FATOR_VARIEDADE': fator_variedade,
        'FATOR_ATENDIMENTO': fator_atendimento, 'FATOR_ESTACIONAMENTO': fator_estacionamento,
        'FUNC_COMPARACAO': func_comparacao, 'FUNC_DISTANCIA': func_distancia,
        'FUNC_ALERTAS': func_alertas, 'FUNC_LISTA': func_lista,
        'FUNC_SUGESTAO': func_sugestao, 'FUNC_HISTORICO': func_historico,
    }
    row_df = pd.DataFrame([row])
    row_df['score_geral']         = row_df[colunas_fatores + colunas_funcoes].mean(axis=1)
    row_df['score_interesse_app'] = row_df[colunas_funcoes].mean(axis=1)

    for col in features_cat:
        val = row_df[col].values[0]
        try:    row_df[col + '_enc'] = le_dict[col].transform([val])[0]
        except: row_df[col + '_enc'] = 0

    X_novo = row_df[features_enc]
    p_usa  = clf_app.predict_proba(X_novo)[0][1]
    p_paga = clf_premium.predict_proba(X_novo)[0][1]
    v_paga = reg.predict(X_novo)[0]

    print('=' * 58)
    print('RESULTADO DA PREDIÇÃO')
    print('=' * 58)
    print(f'Probabilidade de USAR o app:      {p_usa*100:5.1f}%')
    print(f'Probabilidade de PAGAR premium:   {p_paga*100:5.1f}%')
    print(f'Valor previsto que pagaria:        R$ {v_paga:.2f}')
    print()
    print(f'  Perfil: {genero} | {faixa_etaria} | {faixa_salarial}')
    print(f'  Score de interesse no app: {row_df["score_interesse_app"].values[0]:.2f} / 5')
    print('=' * 58)
```

###A lógica de uso do aplicativo é diferente da lógica de compra. Enquanto o engajamento gratuito é movido pela necessidade racional de economizar, a conversão para o Premium é ditada por fatores demográficos como idade e renda superior.

###O modelo identifica que usuários com maior poder aquisitivo (Perfil 2) possuem uma probabilidade de pagar muito maior, mesmo demonstrando menos interesse nas funções básicas de comparação. Isso ocorre porque esse público não busca o app para economizar centavos, mas sim para comprar conveniência e tempo. Em contrapartida, os usuários mais ativos (Perfil 3) usam a ferramenta para reduzir gastos, o que torna o pagamento de uma assinatura uma barreira lógica para o seu objetivo principal de economia.
